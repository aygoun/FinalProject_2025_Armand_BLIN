```python
# Imports
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")
# Constants
RAW_DATA_DIR = "../data/raw/"

# Set pandas display options to show all columns
pd.set_option("display.max_columns", None)
pd.set_option("display.width", None)
pd.set_option("display.max_colwidth", None)
```

```python
# Loading data from Kuairec
interaction_matrix_df = pd.read_csv(RAW_DATA_DIR + "big_matrix.csv")
display(interaction_matrix_df.head())

item_categories_df = pd.read_csv(RAW_DATA_DIR + "item_categories.csv")
item_categories_df["category_features"] = item_categories_df["feat"].map(eval)
display(item_categories_df.head())

item_daily_features_df = pd.read_csv(RAW_DATA_DIR + "item_daily_features.csv")
display(item_daily_features_df.head())

social_network_df = pd.read_csv(RAW_DATA_DIR + "social_network.csv")
social_network_df["friend_list"] = social_network_df["friend_list"].map(eval)
display(social_network_df.head())

user_features_df = pd.read_csv(RAW_DATA_DIR + "user_features.csv")
display(user_features_df.head())
```

```python
# Merge user features
merged = interaction_matrix_df.merge(user_features_df, on="user_id", how="left")

# Merge item categories
merged = merged.merge(item_categories_df, on="video_id", how="left")

# Merge item daily features
item_daily_agg = (
    item_daily_features_df.groupby("video_id").mean(numeric_only=True).reset_index()
)
merged = merged.merge(item_daily_agg, on="video_id", how="left")

# Merge social network info
social_network_df["num_friends"] = social_network_df["friend_list"].apply(len)


merged = merged.merge(
    social_network_df[["user_id", "num_friends"]], on="user_id", how="left"
)
merged["num_friends"] = merged["num_friends"].fillna(0)

# Converting time to datetime
merged["time"] = pd.to_datetime(merged["time"])

display(merged.head())
```

```python
# Stats:
print(f"Number of unique users in merged matrix: {merged['user_id'].nunique()}")
print(f"Number of unique videos in merged matrix: {merged['video_id'].nunique()}")
print(f"Mean watch ratio: {merged['watch_ratio'].mean()}")
print(f"Median watch ratio: {merged['watch_ratio'].median()}")
print("merged.shape:", merged.shape)
print("merged.dtypes:", merged.dtypes)
print("Number of missing values in merged:", merged.isnull().sum())

# Compute average engagement metrics per video
avg_likes = merged['like_cnt'].mean()
avg_comments = merged['comment_cnt'].mean()
avg_shares = merged['share_cnt'].mean()

print(f"Average likes per video: {avg_likes:.0f}")
print(f"Average comments per video: {avg_comments:.0f}")
print(f"Average shares per video: {avg_shares:.0f}")

merged.head()
```

```python
# Let's remove onehot_feat* columns because it's an encrypted column and we don't need it for our analysis
merged.drop(columns=[col for col in merged.columns if col.startswith("onehot_feat")], inplace=True)
```

```python
# Get counts per user
user_counts = merged["user_id"].value_counts()

# Create histogram with log scale
plt.figure(figsize=(12, 6))
plt.hist(user_counts, bins=50, log=True)
plt.title("Distribution of Interactions per User (Log Scale)")
plt.xlabel("Number of Interactions")
plt.ylabel("Count of Users (Log Scale)")
plt.grid(True, alpha=0.3)
plt.show()

# Also plot regular scale for comparison
plt.figure(figsize=(12, 6))
plt.hist(user_counts, bins=50)
plt.title("Distribution of Interactions per User")
plt.xlabel("Number of Interactions")
plt.ylabel("Count of Users")
plt.grid(True, alpha=0.3)
plt.show()
```

```python
# Check for missing values
missing = merged.isnull().mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=missing.index[:20], y=missing.values[:20])
plt.title("Top 20 Cols by Fraction of Missing Values")
plt.ylabel("Missing values %")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(merged["watch_ratio"].dropna(), bins=50, kde=True)
plt.title("Distribution of Watch Ratio")
plt.xlabel("Watch Ratio")
plt.ylabel("Frequency")
plt.xlim(0, 5)  # Limit x-axis to show meaningful range for ratio
plt.grid(True, alpha=0.3)
plt.show()
```

```python
video_inter_counts = merged["video_id"].value_counts()
sns.histplot(video_inter_counts, bins=50, log=True)
plt.title("Distribution of Number of Interactions per Video")
plt.xlabel("Interactions per Video")
plt.ylabel("Number of Videos (log scale)")
plt.show()
```

```python
# User activity degree distribution
plt.figure(figsize=(8, 4))
sns.countplot(
    data=merged,
    x="user_active_degree",
    order=merged["user_active_degree"].value_counts().index,
)
plt.title("User Active Degree Distribution")
plt.xlabel("User Active Degree")
plt.ylabel("Count")
plt.show()

# Number of friends distribution
plt.figure(figsize=(8, 4))
sns.barplot(x=merged["num_friends"].value_counts().index, 
            y=merged["num_friends"].value_counts().values)
plt.title("Number of Friends per User")
plt.xlabel("Number of Friends")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()

# Correlation heatmap for selected numerical features
num_cols = [
    "watch_ratio",
    "play_duration_x",
    "video_duration_x",
    "like_cnt",
    "comment_cnt",
    "share_cnt",
    "num_friends",
]
plt.figure(figsize=(10, 8))
sns.heatmap(merged[num_cols].corr(), annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap (Selected Features)")
plt.show()
```

### Interpretation
- Most users don't have friends on the platform => collaborative-filltering seams not a good way

```python
plt.figure(figsize=(8, 4))
sns.scatterplot(
    x="video_duration_x",
    y="watch_ratio",
    data=merged.sample(5000, random_state=42),
    alpha=0.3,
)
plt.title("Watch Ratio vs Video Duration")
plt.xlabel("Video Duration (s)")
plt.ylabel("Watch Ratio")
plt.show()

plt.figure(figsize=(8, 4))
sns.boxplot(x="user_active_degree", y="watch_ratio", data=merged)
plt.title("User Active Degree vs Watch Ratio")
plt.xlabel("User Active Degree")
plt.ylabel("Watch Ratio")
plt.show()

# Video duration distribution
plt.figure(figsize=(8, 4))
sns.histplot(merged["video_duration_x"].dropna(), bins=50, kde=True)
plt.title("Distribution of Video Duration (x)")
plt.xlabel("Video Duration (seconds)")
plt.ylabel("Frequency")
plt.show()
```

# Interpretation:
- The longer the video, the lower is the watch ratio is
- Most video are less than 30000s


```python
# Correlation between likes and watch duration
plt.figure(figsize=(8, 4))
sns.scatterplot(
    x="like_cnt",
    y="video_duration_x",
    data=merged.sample(5000, random_state=42),
    alpha=0.3,
)
plt.title("Video Duration vs Likes")
plt.xlabel("Likes")
plt.ylabel("Video Duration in s")
plt.show()
```

```python
print("Watch ratio > 1:", (merged["watch_ratio"] > 1).sum())
print("Zero-duration videos:", (merged["video_duration_x"] <= 0).sum())
```

```python
# Calculate mean watch ratio per user and select top 20
top_users = (
    merged.groupby("user_id")["watch_ratio"]
    .mean()
    .sort_values(ascending=False)
    .head(20)
)

plt.figure(figsize=(12, 8))
top_users.sort_values().plot(
    kind="barh", color="skyblue"
)

plt.title("Top 20 Users by Avg Watch Ratio")
plt.xlabel("Avg Watch Ratio")
plt.ylabel("User ID")
plt.tight_layout()
plt.show()
```

```python
# Let's get some stats about category_features
merged["category_features"].apply(lambda x: len(x)).value_counts().plot(kind="bar")
plt.title("Distribution of Number of Category Features")
plt.xlabel("Number of Category Features")
plt.ylabel("Count")
plt.show()


# Number of videos per category
merged["category_features"] = merged["category_features"].apply(
    lambda x: eval(x) if isinstance(x, str) else x
)
df_exploded = merged.explode("category_features")

category_counts = df_exploded["category_features"].value_counts().sort_index()
print(category_counts)

plt.figure(figsize=(12, 6))
sns.barplot(
    x=category_counts.index.astype(str), y=category_counts.values, palette="viridis"
)
plt.title("Distribution of Videos per Category")
plt.xlabel("Category ID")
plt.ylabel("Number of Videos")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Number of NaN or [] in category_features
print("Number of NaN in category_features:", merged["category_features"].isna().sum())
print("Number of [] in category_features:", merged["category_features"].apply(lambda x: len(x) == 0).sum())
```

- Liked videos more likely to be shared
- Longer videos less likely to be watched fully
- Social connections not directly linked to engagement => not enough friend per user to be useful
